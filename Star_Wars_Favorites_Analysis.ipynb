```python
from decimal import Decimal
character = ["Han_Solo", "Luke_Skywalker", "Princess_Leia", 
    "Anakin_Skywalker", "Obi_Wan_Kenobi", "Emperor_Palpatine", "Darth_Vader",
    "Lando_Calrissian", "Boba_Fett", "C3P0", "R2D2", "Jar_Jar_Binks", 
    "Padme_Amidala", "Yoda"]   
    
```

# Percentage of Favorable views

```python
favorable_dict = {}
for r in range(0,14):
    x = star_wars.iloc[:,r+15].value_counts()["Very favorably"]
    y = star_wars.iloc[:,r+15].value_counts()["Somewhat favorably"]
    
    #Subtracting 'unfamiliar (N/A)'n responses for percentage calculation
    total_knowing = sum(star_wars.iloc[:,r+15].value_counts()) 
    - star_wars.iloc[:,r+15].value_counts()["Unfamiliar (N/A)"]
    percent_fav = ((x+y)/total_knowing)*100
    
    #rounding to two decimal points
    percent_fav = round(Decimal(percent_fav),2)
    favorable_dict[character[r]] = percent_fav

plt.barh(range(len(favorable_dict)), list(favorable_dict.values()), align = 'center')
plt.yticks(range(len(favorable_dict)), list(favorable_dict.keys()))
plt.spines['bottom'].set_visible\(False)
plt.show()
```

Luke Skywalker is the winner followed closely by Han Solo, Princess Leia, Obi Wan and Yoda

# Percentage of Unfavorable Views

```python
unfavorable_dict = {}
for r in range(0,14):
    x = star_wars.iloc[:,r+15].value_counts()["Very unfavorably"]
    y = star_wars.iloc[:,r+15].value_counts()["Somewhat unfavorably"]
    
    #Subtracting 'unfamiliar (N/A)'n responses for percentage calculation
    total_knowing = sum(star_wars.iloc[:,r+15].value_counts()) 
    - star_wars.iloc[:,r+15].value_counts()["Unfamiliar (N/A)"]
    percent_unfav = ((x+y)/total_knowing)*100
    
    #rounding to two decimal points
    percent_unfav = round(Decimal(percent_unfav),2)
    unfavorable_dict[character[r]] = percent_unfav  
    
plt.barh(range(len(unfavorable_dict)), list(unfavorable_dict.values()), align = 'center')
plt.yticks(range(len(unfavorable_dict)), list(unfavorable_dict.keys()))
plt.show()
```

Jar Jar Binks is the most unfavorably viewed character followed by Darth Vader

# Percentage of Neutral Views

```python
neutral_dict = {}
for r in range(0,14):
    x = star_wars.iloc[:,r+15].value_counts()["Neither favorably nor unfavorably (neutral)"]
    
    #Subtracting 'unfamiliar (N/A)'n responses for percentage calculation
    total_knowing = sum(star_wars.iloc[:,r+15].value_counts()) 
    - star_wars.iloc[:,r+15].value_counts()["Unfamiliar (N/A)"]
    percent_neutral = (x/total_knowing)*100
    
    #rounding to two decimal points
    percent_neutral = round(Decimal(percent_neutral),2)
    neutral_dict[character[r]] = percent_neutral

plt.barh(range(len(neutral_dict)), list(neutral_dict.values()), align = 'center')
plt.yticks(range(len(neutral_dict)), list(neutral_dict.keys()))
plt.show()
```

Looks like the reason we didn't see Boba Fett above is because everyone is kinda neutral to him'

# Unfamiliar with the Character

```python
  

    
unfamiliar_dict = {}
for r in range(0,14):
    #Using total answered responses for percentage calculation
    total_answered = sum(star_wars.iloc[:,r+15].value_counts()) 
    x = star_wars.iloc[:,r+15].value_counts()["Unfamiliar (N/A)"]
    percent_unfam = ((x)/total_answered)*100
    
    #rounding to two decimal points
    percent_unfam = round(Decimal(percent_unfam),2)
    unfamiliar_dict[character[r]] = percent_unfam  
    
plt.barh(range(len(unfamiliar_dict)), list(unfamiliar_dict.values()), align = 'center')
plt.yticks(range(len(unfamiliar_dict)), list(unfamiliar_dict.keys()))
plt.show()
```

Padme Amidala is the most unfamiliar character followed by Emperor Palpatine and Lando
